# 代码实践

```python
# import torch
# import matplotlib.pyplot as plt

def Produce_X(x):
    x0 = torch.ones(x.numpy().size)
    X = torch.stack((x, x0), dim=1)  # 将向量x和x0合并成为一个矩阵X
    return X


x = torch.Tensor([1.4, 5, 11, 16, 21])
y = torch.Tensor([14.4, 29.6, 62, 85.5, 113.4])
X = Produce_X(x)

print(x)
print(y)
print(X)


inputs = X
target = y
w = torch.rand(2, requires_grad=True)  
# y=kx+b,用w1代替k，用w0代替b：y=w1x+w0
# 随机初始化参数向量w，属于叶子节点，需进行自动微分并利用梯度下降法更新
```

```python
def train(epochs=1, learning_rate=0.01):
    
    for epoch in range(epochs):
        # 前向传播
        output = inputs.mv(w)  # 进行矩阵乘向量运算
        loss = (output - target).pow(2).sum()  # 计算损失
        
        # 反向传播
        loss.backward()  # 自动求损失函数关于w的梯度向量
        w.data -= learning_rate * w.grad  # 更新参数
        w.grad.zero_()   # 值得注意的是更新完w后，必须清空梯度值
        if epoch % 80 == 0:
            draw(output, loss)
            
            
    return w, loss


def draw(output, loss):
    plt.cla()  # 清空画布
    plt.scatter(x.numpy(), y.numpy())  # 绘制样本点散点图
    plt.plot(x.numpy(), output.data.numpy(), 'r-', lw=5)  # 绘制回归直线
    plt.text(0.5, 0, 'loss=%s' % (loss.item()), fontdict={'size':20, 'color':'red'})  # 打印loss值
    plt.pause(0.005)
    
    
w, loss = train(10000, learning_rate = 1e-4)
print("final loss:", loss.item())
print("weight:", w.data)
```
